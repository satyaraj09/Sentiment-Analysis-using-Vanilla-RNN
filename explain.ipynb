## 1. Why ANN (Feedforward Neural Network) is not enough

**ANNs assume inputs are independent.**

- They process data **all at once**
- No memory of past inputs
- Order of data does **not** matter

üî¥ **Problem:**
If you change the order of inputs (like words in a sentence), an ANN treats it the same.

**Example:**

- ‚ÄúI am happy‚Äù
- ‚ÄúHappy am I‚Äù

An ANN sees them as nearly identical ‚Üí **loses meaning**

## 2. Why CNN is not enough

**CNNs are great for spatial patterns, not temporal ones.**

- Designed for images (height √ó width)
- Local pattern detection using filters
- Limited memory (fixed window size)

üî¥ **Problem:**
CNNs do not remember **long-term dependencies**.

**Example:**
In a sentence:

> ‚ÄúI grew up in France‚Ä¶ I speak fluent \_\_\_‚Äù

CNN struggles to connect ‚ÄúFrance‚Äù with ‚ÄúFrench‚Äù when they are far apart.

## 3. Why we need RNN

**RNNs are designed for sequential data.**

‚úî They have **memory**
‚úî Output depends on **previous inputs**
‚úî Order of data **matters**

RNN processes data **step-by-step**, passing information forward.

**Mathematically:**
[
h_t = f(x_t, h_{t-1})
]

Where:

- (x_t) = current input
- (h\_{t-1}) = previous memory
- (h_t) = current memory

---

## 4. Problems RNN solves better

RNNs are used when **context matters**:

| Task                   | Why RNN                                |
| ---------------------- | -------------------------------------- |
| Speech recognition     | Sound depends on previous sounds       |
| Language translation   | Word meaning depends on previous words |
| Text generation        | Needs sentence history                 |
| Time-series prediction | Past values affect future values       |
| Chatbots               | Need conversation context              |
